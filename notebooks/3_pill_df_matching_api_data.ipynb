```python
codes = []
names = []
fronts = []
backs = []
shapes = []
colors = []
urls = []

for val in available_list:
  code = data[data["품목일련번호"] == val]["품목일련번호"].tolist()[0]
  name = data[data["품목일련번호"] == val]["품목명"].tolist()[0]
  front = data[data["품목일련번호"] == val]["표시앞"].tolist()[0]
  back = data[data["품목일련번호"] == val]["표시뒤"].tolist()[0]
  shape = data[data["품목일련번호"] == val]["의약품제형"].tolist()[0]
  color = data[data["품목일련번호"] == val]["색상앞"].tolist()[0]
  url = data[data["품목일련번호"] == val]["큰제품이미지"].tolist()[0]

  codes.append(code)
  names.append(name)
  fronts.append(front)
  backs.append(back)
  shapes.append(shape)
  colors.append(color)
  urls.append(url)

final_df = pd.DataFrame({'품목일련번호': codes, '품목명':names, '표시앞':fronts, '표시뒤':backs, '의약품제형':shapes, '색상':colors, '큰제품이미지':urls})
```
